```python
frequency=df.groupby('CustomerID')['TransactionDate'].count().reset_index().rename(columns={'TransactionDate':'Frequency'})
frequency.head()
```

```python
monetary=df.groupby('CustomerID')['TransactionAmount'].sum().reset_index().rename(columns={'TransactionAmount':'Monetary'})
monetary
```

```python
# Step 4: Build df_rfm by merging Recency, Frequency, and Monetary
df_rfm = customer_data[['CustomerID', 'Recency']].copy()
df_rfm = df_rfm.merge(frequency, on='CustomerID', how='left')
df_rfm = df_rfm.merge(monetary, on='CustomerID', how='left')
df_rfm.head()
```

## Step 5: RFM Scoring using Quantiles

```python
# R_Score: lower recency = higher score (5 is best)
df_rfm['R_Score'] = pd.qcut(df_rfm['Recency'], q=5, labels=[5,4,3,2,1]).astype(int)

# F_Score: higher frequency = higher score (use rank to handle ties)
df_rfm['F_Score'] = pd.qcut(df_rfm['Frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5]).astype(int)

# M_Score: higher monetary = higher score
df_rfm['M_Score'] = pd.qcut(df_rfm['Monetary'], q=5, labels=[1,2,3,4,5]).astype(int)

df_rfm[['CustomerID','Recency','Frequency','Monetary','R_Score','F_Score','M_Score']].head(10)
```

## Step 6: Create Combined RFM Segment String

```python
# Concatenate the 3 scores into a single string e.g. '555'
df_rfm['RFM_Segment'] = (
    df_rfm['R_Score'].astype(str) +
    df_rfm['F_Score'].astype(str) +
    df_rfm['M_Score'].astype(str)
)
df_rfm['RFM_Score'] = df_rfm['R_Score'] + df_rfm['F_Score'] + df_rfm['M_Score']
print(df_rfm['RFM_Segment'].value_counts().head(10))
df_rfm[['CustomerID','RFM_Segment','RFM_Score']].head()
```

## Step 7: Assign Business Segment Labels

```python
def assign_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif f >= 4 and r >= 2:
        return 'Loyal Customers'
    elif r >= 4 and 2 <= f <= 3:
        return 'Potential Loyalist'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif m >= 4 and 2 <= f <= 3 and r >= 3:
        return 'Big Spenders'
    elif r == 1 and f <= 2 and m <= 2:
        return 'Lost'
    else:
        return 'Others'

df_rfm['Segment'] = df_rfm.apply(assign_segment, axis=1)
print(df_rfm['Segment'].value_counts())
df_rfm.head(10)
```

## Step 8: Visualizations

### 8.1 Count of Customers in Each Segment

```python
segment_counts = df_rfm['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Count']

plt.figure(figsize=(10, 6))
colors = sns.color_palette('Set2', len(segment_counts))
bars = plt.bar(segment_counts['Segment'], segment_counts['Count'], color=colors, edgecolor='black')
for bar, count in zip(bars, segment_counts['Count']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 3,
             str(count), ha='center', va='bottom', fontweight='bold')
plt.title('Number of Customers per Segment', fontsize=14, fontweight='bold')
plt.xlabel('Segment')
plt.ylabel('Customer Count')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

### 8.2 Revenue Contribution per Segment

```python
seg_revenue = df_rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False).reset_index()
seg_revenue.columns = ['Segment', 'TotalRevenue']
seg_revenue['Revenue_Pct'] = (seg_revenue['TotalRevenue'] / seg_revenue['TotalRevenue'].sum() * 100).round(2)

fig, ax = plt.subplots(figsize=(10, 6))
colors = sns.color_palette('tab10', len(seg_revenue))
bars = ax.bar(seg_revenue['Segment'], seg_revenue['TotalRevenue'], color=colors, edgecolor='black')
for bar, pct in zip(bars, seg_revenue['Revenue_Pct']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20000,
            f'{pct}%', ha='center', va='bottom', fontsize=9, fontweight='bold')
ax.set_title('Revenue Contribution per Customer Segment', fontsize=14, fontweight='bold')
ax.set_xlabel('Segment')
ax.set_ylabel('Total Revenue (â¹)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'â¹{x/1e6:.1f}M'))
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
print(seg_revenue)
```

### 8.3 Recency vs Monetary Scatter Plot (Colored by Segment)

```python
plt.figure(figsize=(12, 7))
segments = df_rfm['Segment'].unique()
palette = sns.color_palette('tab10', len(segments))
seg_color = dict(zip(segments, palette))

for seg in segments:
    subset = df_rfm[df_rfm['Segment'] == seg]
    plt.scatter(subset['Recency'], subset['Monetary'], label=seg,
                alpha=0.6, s=60, color=seg_color[seg])

plt.title('Recency vs Monetary Value by Segment', fontsize=14, fontweight='bold')
plt.xlabel('Recency (Days since last purchase)')
plt.ylabel('Monetary Value (â¹)')
plt.legend(title='Segment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 8.4 Pareto Analysis: Top 20% Customers â 80% Revenue

```python
# Sort customers by Monetary descending
df_pareto = df_rfm[['CustomerID', 'Monetary']].sort_values('Monetary', ascending=False).reset_index(drop=True)
df_pareto['CumulativeRevenue'] = df_pareto['Monetary'].cumsum()
df_pareto['CumulativeRevenuePct'] = df_pareto['CumulativeRevenue'] / df_pareto['Monetary'].sum() * 100
df_pareto['CustomerPct'] = (df_pareto.index + 1) / len(df_pareto) * 100

# Find 20% customer cutoff
top20_rev = df_pareto[df_pareto['CustomerPct'] <= 20]['CumulativeRevenuePct'].max()
print(f'Top 20% customers contribute {top20_rev:.1f}% of total revenue')

fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.bar(df_pareto['CustomerPct'], df_pareto['Monetary'],
        width=0.1, color='steelblue', alpha=0.5, label='Individual Revenue')
ax1.set_xlabel('Customer Percentile (%)')
ax1.set_ylabel('Individual Revenue (â¹)', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')

ax2 = ax1.twinx()
ax2.plot(df_pareto['CustomerPct'], df_pareto['CumulativeRevenuePct'],
         color='crimson', linewidth=2.5, label='Cumulative Revenue %')
ax2.axvline(x=20, color='orange', linestyle='--', linewidth=1.5, label='20% Customer Mark')
ax2.axhline(y=top20_rev, color='green', linestyle='--', linewidth=1.5, label=f'{top20_rev:.1f}% Revenue')
ax2.set_ylabel('Cumulative Revenue (%)', color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='center right')
plt.title(f'Pareto Analysis: Top 20% Customers Drive {top20_rev:.1f}% of Revenue', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
